```python
import pandas as pd

data = {
    'Piloto': ['Hamilton', 'Vettel', 'Raikkonen', 'Verstappen', 'Bottas', 'Ricciardo'],
    'Escuderia': ['Mercedes', 'Ferrari', 'Ferrari', 'Red Bull', 'Mercedes', 'Red Bull'],
    'Puntos': [408, 320, 251, 249, 247, 170],
    'Victorias': [11, 5, 1, 2, 0, 2]
}

df = pd.DataFrame(data)

solo_ferrari = df[df['Escuderia'] == 'Ferrari']

total_victorias = df['Victorias'].sum()

puntos_constructores = df.groupby('Escuderia')['Puntos'].sum()

print(puntos_constructores)
```

```python
puntos_constructores.plot(kind='bar', color=['red', 'silver', 'blue'], title="Puntos Totales F1 2018")
```
