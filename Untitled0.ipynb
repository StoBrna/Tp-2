```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

null = df['Age'].isnull().sum()
print(f'Pasajeros sin edad: {null}')

sns.histplot(df['Age'].dropna(), kde = True)
plt.show()
```

```python
promedy = df['Age'].mean()
median = df['Age'].median()

print(f'Edad Promedio: {promedy:.2f} a単os')
print(f'Edad Mediana: {median} a単os')
```

```python
df['Age_promedy'] = df['Age'].fillna(promedy)
df['Age_median'] = df['Age'].fillna(median)

print('Nulos en Age_promedy: ', df['Age_promedy'].isnull().sum())
```

```python
plt.figure(figsize = (10, 5))
sns.kdeplot(df['Age'], label = 'Original', fill = True, alpha = 0.3)
sns.kdeplot(df['Age_promedy'], label = 'Promedio', fill = True, alpha = 0.3)
sns.kdeplot(df['Age_median'], label = 'Mediana', fill = True, alpha = 0.3)
plt.legend()
plt.title('Efecto de imputar datos: ')
plt.show()
```

```python
print("Promedio de edad por clase:")
print(df.groupby('Pclass')['Age'].mean())

df['Age_Clase'] = df['Age'].fillna(df.groupby('Pclass')['Age'].transform('mean'))

plt.figure(figsize=(10, 5))
sns.kdeplot(df['Age_Promedio'], label='Relleno Bruto (Promedio General)', color='orange')
sns.kdeplot(df['Age_Clase'], label='Relleno Inteligente (Por Clase)', color='green')
plt.legend()
plt.title("Comparaci坦n: Relleno General vs. Por Clase Social")
plt.show()
```

```python
print("Promedio de edad por clase:")
print(df.groupby('Pclass')['Age'].mean())

df['Age_Clase'] = df['Age'].fillna(df.groupby('Pclass')['Age'].transform('mean'))

plt.figure(figsize=(10, 5))
sns.kdeplot(df['Age'], label='Original (Real)', fill=True, alpha=0.3, color='blue')
sns.kdeplot(df['Age_promedy'], label='Mal: Promedio General', color='orange') # Tu variable se llama 'promedy'
sns.kdeplot(df['Age_Clase'], label='Bien: Por Clase', color='green', linewidth=3)

plt.legend()
plt.title("Comparaci坦n Final: La l鱈nea verde deber鱈a seguir mejor a la azul")
plt.show()
```
