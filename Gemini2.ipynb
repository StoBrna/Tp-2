```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df_cars = pd.read_csv(url)

print("===== VISTA PREVIA DEL DATASET ====")
print(df_cars.head())
print("\n" + "="*50 + "\n")

top_power = df_cars.sort_values(by = "horsepower", ascending = False).head(5)

print ("===== TOP 5 AUTOS MAS POTENTES ====")
print(top_power[["name", "horsepower", "cylinders"]])
print("\n" + "="*50 + "\n")

cars_usa_v8 = df_cars[(df_cars["origin"] == "usa") & (df_cars["cylinders"] == 8)]

print(f"===== CANTIDAD DE V8 AMERICANOS ENCONTRADOS: {len(cars_usa_v8)}=====")
print("EJEMPLOS: ")
print(cars_usa_v8[["name", "mpg"]].head(5))
```

```python
df_cars.plot(kind = 'scatter', x = 'horsepower', y = 'mpg', c = 'cylinders', colormap = 'viridis', title = "Potencia vs. Consumo")
```

```python
lb_cars = df_cars[df_cars['cylinders'] == 6].sort_values(by='mpg', ascending=False)

print("===== EL REY DE LOS 6 CILINDROS (CELESTE) =====")
print(lb_cars[['name', 'horsepower', 'mpg']].head(1))
```
